```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
treino = pd.read_csv('train.csv')
treino
```

```python
teste = pd.read_csv('test.csv')
teste
```

```python
base_completa = pd.concat([treino, teste])
base_completa
```

```python
base_completa.describe()
```

```python
base_completa.drop(['Name', 'Ticket', 'Cabin'], axis=1, inplace=True)
```

```python
base_completa
```

```python
base_completa.isnull().sum()
```

```python
base_completa['Age'].fillna(base_completa['Age'].mean(), inplace=True)
```

```python
base_completa['Fare'].fillna(base_completa['Fare'].mean(), inplace=True)
```

```python
base_completa['Embarked'].value_counts()
```

```python
base_completa['Embarked'].fillna('S', inplace=True)
```

```python
base_completa.isnull().sum()
```

```python
treino = base_completa[base_completa['Survived'].notnull()]
teste = base_completa[base_completa['Survived'].isnull()]
```

```python
treino
```

```python
teste
```

```python
X_treino = treino.drop(['Survived'], axis = 1)
y_treino = treino['Survived']
X_teste = teste.drop(['Survived'], axis = 1)
y_teste = teste['Survived']
```

```python
y_treino = y_treino.astype(int)
y_treino
```

```python
X_treino
```

```python
X_teste
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
X_treino['Sex'] = label_encoder.fit_transform(X_treino['Sex'])
X_treino['Embarked'] = label_encoder.fit_transform(X_treino['Embarked'])
X_teste['Sex'] = label_encoder.fit_transform(X_teste['Sex'])
X_teste['Embarked'] = label_encoder.fit_transform(X_teste['Embarked'])
```

```python
X_treino
```

```python
X_teste
```

```python
y_treino.value_counts()
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(sampling_strategy='minority')
X_treino_over, y_treino_over = smote.fit_resample(X_treino, y_treino)
```

```python
y_treino_over.value_counts()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_treino_over = scaler.fit_transform(X_treino_over)
X_teste = scaler.fit_transform(X_teste)
```

```python
from sklearn.linear_model import LogisticRegression
logistic_regression = LogisticRegression()
logistic_regression.fit(X_treino_over, y_treino_over)
```

```python
previsoes = logistic_regression.predict(X_teste)
```

```python
sub = pd.Series(previsoes, index=teste['PassengerId'], name='Survived')
sub.shape
```

```python
sub.to_csv('previsoes.csv', header=True)
```
